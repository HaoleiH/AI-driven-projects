```python
import google.generativeai as genai
from IPython.display import Markdown
import PIL.Image
from IPython.display import Image
```

```python
from google.colab import userdata
GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')

genai.configure(api_key=GOOGLE_API_KEY)
```

# first example

## image

```python


sample_file = PIL.Image.open('/content/test1.png')

# prompt: display /content/test1.png

from IPython.display import Image
Image('/content/test1.png')
```

## response

```python

# Choose a Gemini model.
model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest")

# Create a prompt.
prompt = "extract data to plot the cruve in the graph, show them in two columns, first column showing x coordinate, second column showing y coordinate.return only csv file"

response = model.generate_content([sample_file, prompt])

Markdown(response.text)
```

## plot extracted data

```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt

# Assuming 'response.text' contains the CSV data
csv_data = response.text

# Use StringIO to treat the string as a file-like object
df = pd.read_csv(StringIO(csv_data[6:-3]),names=["x","y"])

plt.plot(df["x"],df["y"],"*-")
```

## response

```python

# Choose a Gemini model.
model = genai.GenerativeModel(model_name="gemini-1.5-pro-latest")

# Create a prompt.
prompt = "extract data to plot the cruve in the graph, show them in two columns, first column showing x coordinate, second column showing y coordinate, the step in x coordiante is 0.05. return only csv file"

response = model.generate_content([sample_file, prompt])

Markdown(response.text)
```
