```python
# prompt: plot file at https://github.com/HaoleiH/AI-driven-projects/blob/main/example-data/sample-PLdata.txt, fit data in x range 500 to 540 to a gaussian function, show the fitting parameters

import pandas as pd
import plotly.express as px
import requests
from io import StringIO
import numpy as np
from scipy.optimize import curve_fit

# Download the file from GitHub
url = "https://raw.githubusercontent.com/HaoleiH/AI-driven-projects/main/example-data/sample-PLdata.txt"
response = requests.get(url)
data = response.text

# Read the data into a pandas DataFrame
df = pd.read_csv(StringIO(data), delimiter='\t')

# Define the Gaussian function
def gaussian(x, amplitude, mean, stddev):
    return amplitude * np.exp(-((x - mean) / 4 / stddev)**2)

# Select data within the specified x range
x_data = df[df.columns[0]]
y_data = df[df.columns[1]]
mask = (x_data >= 500) & (x_data <= 540)
x_fit = x_data[mask]
y_fit = y_data[mask]


# Fit the data to a Gaussian function
popt, pcov = curve_fit(gaussian, x_fit, y_fit, p0=[1000, 520, 10])  # Initial guess for parameters

# Get the fitting parameters and their uncertainties
amplitude = popt[0]
mean = popt[1]
stddev = popt[2]
amplitude_err = np.sqrt(pcov[0, 0])
mean_err = np.sqrt(pcov[1, 1])
stddev_err = np.sqrt(pcov[2, 2])

# Print the fitting parameters
print(f"Amplitude: {amplitude:.2f}  {amplitude_err:.2f}")
print(f"Mean: {mean:.2f}  {mean_err:.2f}")
print(f"Standard Deviation: {stddev:.2f}  {stddev_err:.2f}")

# Create the plot
fig = px.line(df, x=df.columns[0], y=df.columns[1:],
              title='Plot of Sample PL Data with Gaussian Fit',
              labels={df.columns[0]: 'X-axis', 'value':'Y-axis'},
              range_x=[500, 600],
              range_y=[0, 1500])

# Add the Gaussian fit to the plot
x_fit_plot = np.linspace(500, 540, 100)
y_fit_plot = gaussian(x_fit_plot, *popt)

fig.add_scatter(x=x_fit_plot, y=y_fit_plot, mode='lines', name='Gaussian Fit')

fig.show()
```
